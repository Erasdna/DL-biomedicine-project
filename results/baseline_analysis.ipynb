## Tabula muris - fealstm

```python
results = []

for name in ['base', 'basepp', 'matching', 'proto', 'maml']:

    res = pd.read_csv(f'tbm_{name}_res.csv')
    runs = pd.read_csv(f'tbm_{name}_runs.csv')
    df_baseline = res.merge(runs, left_on='run_name', right_on='Name')

    df_baseline = df_baseline.loc[:, ['split', 'acc_mean', 'acc_std', 'dataset.set_cls.n_support', 'method.cls.n_way']]
    df_baseline_test = df_baseline.loc[df_baseline.split == 'test']
    df_baseline_test['baseline'] = name
    results.append(df_baseline_test)
```

```python
for method in ['feat', 'fealstm', 'feads']:
    res = pd.read_csv(f'tbm_{method}.csv')
    runs = pd.read_csv(f'tbm_{method}_runs.csv')
    df = res.merge(runs, left_on='run_name', right_on='Name')

    df = df.loc[:, ['split',
     'acc_mean',
     'acc_std',
     'run_name','lr','method.cls.n_head',
     'dataset.set_cls.n_support',
     'method.cls.n_way',
     'method.cls.dim_feedforward',
     'method.cls.dropout',
     'method.cls.num_layers',
     'method.cls.score._target_','loss']]

    df_val = df.loc[df.split.isin(['val'])]
    #df_tbm_val.sort_values(by=['acc_mean'], ascending=False)

    best_hyps = df_val.sort_values(by=['acc_mean'], ascending=False).groupby(['dataset.set_cls.n_support','method.cls.n_way', 'split']).first()

    best_runs = df.loc[df.run_name.isin(best_hyps.run_name)]
    test_runs = best_runs.loc[best_runs.split == 'test', ['split', 'acc_mean', 'acc_std', 'dataset.set_cls.n_support', 'method.cls.n_way']]
    test_runs['baseline'] = method
    results.append(test_runs)
```

```python
results = pd.concat(results)
```

```python
results_5_shot = results.loc[results['dataset.set_cls.n_support'] == 5]
results_1_shot = results.loc[results['dataset.set_cls.n_support'] == 1]
```

```python
results_5_shot.to_csv('tbm_base5shot.csv')
results_1_shot.to_csv('tbm_base1shot.csv')
```
